####2. SelectKBest

a. if_classif

```python
# Select KBest for Classification Problems

from sklearn.feature_selection import SelectKBest, f_classif

# Apply SelectKBest
Kbest_classif = SelectKBest(score_func=f_classif, k=6)  # Use ANOVA F-test for classification
Kbest_classif.fit(X_train, y_train)

# Print scores for the features
for i in range(len(Kbest_classif.scores_)):
    print(f'Feature {i} : {round(Kbest_classif.scores_[i], 3)}')

# Visualization of feature scores
import matplotlib.pyplot as plt

plt.bar(
    [X_train.columns[i] for i in range(len(Kbest_classif.scores_))],
    Kbest_classif.scores_
)
plt.xticks(rotation=90)  # Rotate feature names for readability
plt.rcParams["figure.figsize"] = (8, 6)  # Set figure size
plt.show()
```

b. mutual_info_classif

```python
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import SelectKBest

# Apply SelectKBest with mutual_info_classif
infogain_classif = SelectKBest(score_func=mutual_info_classif, k=6)
infogain_classif.fit(X_train, y_train)

# Print scores for the features
for i in range(len(infogain_classif.scores_)):
    print(f'Feature {i} : {round(infogain_classif.scores_[i], 3)}')

# Plot the scores
import matplotlib.pyplot as plt

plt.bar(
    [X_train.columns[i] for i in range(len(infogain_classif.scores_))],
    infogain_classif.scores_
)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.rcParams["figure.figsize"] = (8, 6)  # Set figure size
plt.show()
```

#### 3. SelectPercentile

```python
from sklearn.feature_selection import SelectPercentile, f_classif

# Apply SelectPercentile with a 50% threshold
percentile = SelectPercentile(score_func=f_classif, percentile=50)
percentile.fit(X_train, y_train)

# Check which features are selected
print(percentile.get_support())  # Boolean array indicating selected features

# Display feature names
print(X_train.columns)
```

#### 4. SelectFromModel

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

# Model-based feature selection
model_based_feature = SelectFromModel(RandomForestClassifier(n_estimators=100, random_state=42), threshold="median")
model_based_feature.fit(X_train, y_train)

# Get selected features
print(model_based_feature.get_support())  # Boolean array
print(X_train.columns)  # Feature namesget_support
```

#### 5. Recursive Feature Elimination (RFE)

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

# Recursive Feature Elimination
rfe_features = RFE(RandomForestClassifier(n_estimators=100, random_state=42), n_features_to_select=6)
rfe_features.fit(X_train, y_train)

# Get selected features
print(rfe_features.get_support())  # Boolean array
print(X_train.columns)  # Feature names
```

#### 6. Recursive Feature Elimination with Cross-Validation (RFECV)

```python
from sklearn.feature_selection import RFECV
from sklearn.ensemble import RandomForestClassifier

# Recursive Feature Elimination with Cross-Validation
rfecv = RFECV(estimator=RandomForestClassifier(), step=1, cv=5, scoring='accuracy')  # 5-fold CV
rfecv.fit(X_train, y_train)

# Get the optimal number of features and their names
print(f"Optimal number of features: {rfecv.n_features_}")
print(f"Best features: {X_train.columns[rfecv.support_]}")
```
